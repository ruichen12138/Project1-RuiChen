```python
import pandas as pd

file_path = 'white-wine-price-rating.csv' 

wine_data = pd.read_csv(file_path)

wine_data.head()
```

```python
import pandas as pd
from scipy import stats

x = wine_data['WineRating']  
y = wine_data['WinePrice']

slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

print(f"Correlation Coefficient: {r_value}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

predicted_y = intercept + slope * x

plt.figure(figsize=(10, 6))

plt.scatter(x, y, alpha=0.5, label='Actual WinePrice')

plt.plot(x, predicted_y, color='red', label='Predicted WinePrice')

plt.title('Linear Regression: WineRating vs. WinePrice')
plt.xlabel('WineRating')
plt.ylabel('WinePrice ($)')
plt.legend()

plt.show()
```
